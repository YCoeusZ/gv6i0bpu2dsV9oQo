# Suggested metrics to evaluate the models 

There is no "target" in current context, to evaluate the models, we will need human (or otherwise) outside evaluation, we suggest the following metrics to do so. 

- Recall@k: If we only retrieve the top k documents, what proportion or the total relevant documents have we retrieved. 
- Precision@k: If we only retrieve the top k documents, what proportion of the retrieved are actually relevant. 
- Hit@k: If we only retrieve the top k documents, do we actually retrieve anything relevant? 
- Mean Reciprocal Rank (MRR): Consider the retriever provides a relevancy ranking of documents for a query $q$ in descending order: 
    $$
    d_2, d_3, d_1, d_0\; . 
    $$
    Let the (human selected) actual relevant set of documents be $d_4, d_0, d_3$. 
    We define 
    $$
    rank(q):=\text{ the ranking of the first actually relevant document according to the retriever. }
    $$
    In above example, $rank(q)=2$ which is the ranking of $d_3$ - the first in the ranking according to the retriever that is actually relevant (according to human). 
    Then we define 
    $$
    RR(q):=\begin{cases}
    \frac{1}{rank(q)} & \text{ if a relevant document is retrieved } \\ 
    0 & o.w. 
    \end{cases}\; . 
    $$
    Then the 
    $$
    MRR:= \frac{1}{|Q|}\sum\limits_{q\in Q} RR(q)\; , 
    $$
    where $Q$ is the set of all queries. 

    So, intuitively, higher MRR means that relevant docs appear earlier according to the ranking of the retriever.
